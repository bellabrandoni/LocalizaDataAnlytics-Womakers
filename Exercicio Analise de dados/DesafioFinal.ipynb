
2.  Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mode]


```python

#Obter todos os valores únicos da coluna "Profissão"
valores_profissao = df_sono['Profissão'].unique()
valores_profissao
```

```python
#Cacuçando a média 
# Agrupando por "Profissão" e calculando a média de "Duração de Sono"
media_sono_por_profissao = df_sono.groupby('Profissão')['Duração do sono'].mean()

# Exibindo resultado
print(f' Média de sono por {media_sono_por_profissao}')
```

```python
moda_sono_por_profissao = df_sono.groupby('Profissão')['Duração do sono'].apply(lambda x: x.mode()[0])

print(f"Moda da Duração de Sono por Profissão: {moda_sono_por_profissao}")
```

```python
#Cacuçando a mediana
# Agrupando por "Profissão"e "Duração de Sono" e calculando a mediana 
mediana_sono_por_profissao = df_sono.groupby('Profissão')['Duração do sono'].median()

# Exibindo resultado
print(f'A  Mediana de sono por {mediana_sono_por_profissao}')
```

3. Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?


```python
df_sono.head()
```

```python
# Filtrar os engenheiros de software diretamente na coluna Profissão
eng_s = df_sono.loc[df_sono['Profissão'] == 'Eng. de Software']

# Filtrar os engenheiros de software obesos (Categoria IMC == 'Obeso(a)')
eng_s_obesos = eng_s.loc[eng_s ['Categoria IMC'] == 'Obesidade']

# Calcular o total de engenheiros de software
tot_eng_s = eng_s.shape[0]

# Contador para calcular o total de engenheiros de software obesos
eng_s_obesos = eng_s_obesos.shape[0]

# Verificação para corrigir divisão por zero ao calcular a porcentagem
porcent_eng_obesos = (eng_s_obesos / tot_eng_s) * 100 if tot_eng_s  > 0 else 0

# Exibir a porcentagem
print(f"Porcentagem de engenheiros de software obesos: {porcent_eng_obesos:.1f}%")
```

4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)


```python
df_sono.head()
```

```python
#utilizando isin() para filtrar as profissões de Advogado(a) e Representante de Vendas dentro da coluna Profissão.

filtrando_adv_rep = df_sono[df_sono['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]

# Calculando a média de sono 
media_sono = filtrando_adv_rep['Duração do sono'].mean()

# Calculando a média geral de sono de todas as profissoes
media_sono_geral = df_sono['Duração do sono'].mean()

# Exibir as médias
print(f"Média de sono de Advogados ou Representantes de Vendas: {media_sono:.2f} horas")
print(f"Média geral de sono: {media_sono_geral:.2f} horas")

# Verificando se advogar ou ser representante de vendas faz você dormir menos
if media_sono < media_sono_geral:
    print("Advogar ou ser Representante de Vendas faz você dormir menos.")
else:
    print("Advogar ou ser Representante de Vendas não faz você dormir menos.")
```

5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)


```python
#print(f' Média de sono por {media_sono_por_profissao}')

#Criando variavel para filtrar Medicos e Enfermeiros na coluna profissão
filtrando_med_enf = df_sono[df_sono['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])]

media_sono_med_enf = filtrando_med_enf.groupby('Profissão')['Duração do sono'].mean()
#calculando media de sono das duas profissoes 
media_sono_med_enf

# Verificando  quem dorme menos
if media_sono_med_enf['Enfermeiro(a)'] < media_sono_med_enf['Médico(a)']:
    print("Profissionais de Enfermagem dormem menos.")
else:
    print("Profissionais de Medicina dormem menos.")
```
