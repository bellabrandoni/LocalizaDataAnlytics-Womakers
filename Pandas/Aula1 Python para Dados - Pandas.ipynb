# Agregação de dados

**Utilizando o arquivos: concurso.csv**

7. Qual a média de idade por estado?

```python
#chamando o arquivo csv 
df_concurso = pd.read_csv('dados_concurso.csv')
```

```python
#printando head 
df_concurso.head()
```

```python
#Quantas pessoas são do estado do Piauí?
#selecionando primeiro a tabela, depois selecionando o estado referente a PI e depois selecionando numero de inscrição. Depois fazendo count de tudo isso 
df_concurso[df_concurso['Estado'] == 'PI']['Número de Inscrição'].count()
```

```python
#Qual a média de idade por estado?
#Selecionando na tabela concurso e arupando por numero de inscição e fazendo um count com o a quantidade por região a partir do estado
df_concurso[['Número de Inscrição', 'Estado']].groupby('Estado').count()
```

```python

#Selecionando na tabela as colunas  ['Escolaridade', 'Número de Inscrição', 'Estado'] e agrupando por estado e escolaridade e fazendo um count disso tudo
df_concurso[['Escolaridade', 'Número de Inscrição', 'Estado']].groupby(['Estado', 'Escolaridade']).count()
```
