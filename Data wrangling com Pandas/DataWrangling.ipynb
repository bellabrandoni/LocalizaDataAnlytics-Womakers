## Data wrangling  com Pandas 


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("collected research people.csv")
```

```python
df.head()
```

```python
#conultando coluna 	Unnamed: 0
df['Unnamed: 0']

#excluindo colina Unnamed:

df = df.drop(columns= ['Unnamed: 0'])

```

```python
df.head()
```

```python
#renomeando colunas com dicionario de mapeamento
novas_colunas = {'Endereerço' : 'Endereço',
                 'phone' : 'Telefone',
                 'anual_income' : 'Renda Anual',
                 'AAge' : 'Idade',
                 'n_filhos' : 'N Filhos',
                 'diariamente_sun blocker': 'Usa protetor solar diariamente',
                 'Food_restrictions': 'Restrição alimentar',
                 'streaming': 'Streaming'}

#substituindo os nomes com a função rename e passando colunas recebe novas_colunas
df.rename(columns=novas_colunas, inplace=True)
df.head()
```

```python
#conferindo caracteres especiais na coluna Nome
#df.Nome
#substituindo caracteres especiais na coluna Nome
#fazer manipulacao de string linha por linha com df.Nome.str e replace
#substituindo caracteres especiais com \W e substituindo por nada " "
#df.Nome.str.replace('\W', ' ')

df['Nome'] = df['Nome'].str.replace('[^a-zA-Zà-úÀ-Ú0-9 ]', '', regex=True)
#df.Nome
```

```python
df['Telefone'] = df['Telefone'].str.replace('[^0-9() +-]', '')

df.Telefone
```

```python
df.loc[df.Telefone.str.contains('[*]', na= False), 'Telefone'] = np.nan
```

```python
df.Telefone
```

```python
#Estado civil - fazendo um replace dos termos incompletos com o dicionario 
df['Estado Civil'] = df['Estado Civil'].replace({'S': 'Solteiro',
                            'C': 'Casado',
                            'D':'Divocriado',
                            'V': 'Viúvo' })
```
